```python
#Write a program to take a string as input and print its length.
input = "Swaroop"
print(len(input))
```

```python
#Given a string, convert it to uppercase and lowercase.
input = "heyy there"
print(input.upper())
print(input.lower())
```

```python
#Write a program to count the number of vowels in a string
s = "Swaroop"
vowels = "aeiou"
count = 0

for ch in s:
    if ch in vowels:
        count += 1
print("number of vowels:",count)
```

```python
#Check whether a given string is a palindrome (example: madam).
s = "nitin"
if s == s[::-1]:
    print("palindrome")
else:
    print("not a palindrome")
```

```python
#Given a sentence, count the number of words in it.
sentence = "Learning Python is simple and Python is powerful"
words = sentence.split()

print("number of words:",len(words))
```

```python
#Replace all spaces in a string with underscores (_).
input = "Learning Python is simple and Python is powerful"
result = input.replace(" ","_")

print(result)
```
