```python
print("Check if a number is both even and greater than 10")
number = 18
is_even = number % 2 == 0
is_greater_than_10 = number > 10
result = is_even and is_greater_than_10
print(f"number : {number}")
print(f"is even : {is_even}")
print(f"is greater than 10 : {result}")
```
