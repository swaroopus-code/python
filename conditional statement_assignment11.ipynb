```python
# question 1: Simple Age Check Write a program that checks if a person can get a driver's license (age >= 16).

age = int(input("enter your age: "))
if age >= 16:
    print("Eligible to get license")
else :
    print("Under age")
```

```python
# question 2: Positive or Negative Write a program that checks if a number is positive, negative, or zero.
num = int(input("enter the number: "))
if num > 0:
    print(f"{num} is positive")
elif num < 0:
    print(f"{num} is negative")
else:
    print(f"{num} is zero")
```

```python
# question 3 :  Larger Number Write a program that compares two numbers and prints which one is larger.
num1 = int(input("enter number 1:"))
num2 = int(input("enter number 2:"))
if num1 > num2:
    print(f"{num1} is greater than {num2}")
elif num1 < num2:
    print(f"{num2} is greater than {num1}")
else:
    print("both are equal")
```

```python
# question 4 :  Calculator Discount A store offers:
#10% discount if purchase > $100
#5% discount if purchase > $50
#No discount otherwise
#Calculate the final price.

purchase = int(input("enter the amount: "))
if purchase > 100:
    discount = 0.10 * purchase
    final_price = purchase - discount
    print(discount)
    print("10% discount")
    print('final price:',final_price)
elif purchase > 50:
    discount = 0.05 * purchase
    final_price = purchase - discount
    print(discount)
    print("5% discount")
    print('final price:',final_price)
else:
    print("no discount")
    print('final price:',purchase)
```

```python
# question 5 :Leap Year Checker Write a program to check if a year is a leap year:
# Divisible by 4 AND (not divisible by 100 OR divisible by 400)
year = int(input("enter a year:"))
if year % 4 == 0 and (year % 100 !=0 or year % 400 ==0): #A year is leap if: divisible by 4 not divisible by 100 OR divisible by 400
    print("leap year")
else:
    print("not a leap year")
```

```python
#Grade Calculator with Bonus Calculate grades (A-F) but add +1 bonus grade level if attendance > 90%.
attendance = int(input("enter your attendance:"))
grade = input("enter your grade:(A-F)")
if attendance > 90:
    if grade == 'B':
        grade = 'A'
    elif grade == 'C':
        grade = 'B'
    elif grade == 'D':
        grade = 'C'
    elif grade == 'F':
        grade = 'D'

print("finale grade:",grade)
```

```python
#Login System with Attempts Create a login system that:Checks username and password Tracks login attempts (max 3) Locks account after failed attempts

correct_username = "admin"
correct_password = "1234"

attempts = 0
max_attempts = 3

while attempts < max_attempts:
    username = input("enter username")
    password = input("enter password")

    if username == correct_username and password == correct_password:
        print("login success")
        break
    else:
        attempts += 1
        print("wrong attempt, attempt left: ",max_attempts - attempts)
    
if attempts==max_attempts:
    print("many attempts!!")
```
