```python
# 1.Count how many vowels are in a given string.
string = input("enter the string").lower()
vowels = 'aeiou'
count = 0
for i in string:
    if i in vowels:
        count += 1

print(f'in "{string}" vowels are {count}')
```

```python
# 2.Reverse a number using ONLY a for loop (no slicing, no string reverse).
num = int(input("enter the number to be reversed:"))
reverse = 0
for i in range(len(str(num))):
    last_digit = num % 10 # get last digit
    num = num // 10 # remove last digit
    reverse = reverse * 10 + last_digit
print(reverse)
```

```python
# 3.Find the sum of digits of a number using a loop.
num = int(input("Enter number: "))
total = 0
for i in range(len(str(num))):
    last_digit = num % 10
    total += last_digit
    num = num // 10
print(total)
```

```python
#Question 6:Check whether a number is prime using a for loop
num = 3
is_prime = True
for i in range(2,num):
    num % i == 0
    is_prime = False
    break
if is_prime and num > 1:
    print("its prime")
else:
    print("not prime")
```

```python
# Question 13: Check whether a string is a palindrome
string = input("enter the string:")
reverse = ""
for ch in string:
    reverse = ch + reverse
if string == reverse:
    print("palindrome")
else:
    print("not palindrome")
```

```python
# Question 14 : Find duplicate values in a list.
nums = [1, 2, 3, 2, 4, 5, 1]
duplicate = []
seen = []
for i in nums:
    if i in seen:
        if i not in duplicate:
            duplicate.append(i)
    else:
        seen.append(i)
print("duplicate values:",duplicate)
# seen → stores numbers already visited
# duplicates → stores repeated numbers
# For each element:
#   if element already in seen:
#       it is a duplicate → add to duplicates (only once)
#   else:
#       add element to seen
# This catches duplicates even if they are far apart.
```

```python
# 15. Print the frequency of each character in a string.
# Example: "banana"

string = "banana"
freq = {}
for ch in string:
    if ch == string:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```
