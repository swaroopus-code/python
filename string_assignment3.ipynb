```python
# Question 10: Remove all vowels from "Computer Science"

vowels = "aeiouAEIOU"
word = "Computer Science"

result = ""
for ch in word:
    if ch not in vowels:
        result += ch
print(result)
```

```python
# Question 11: Find the most frequent character in "mississippi"
#
char = "mississippi"
freq = {}
for ch in char:
    freq[ch] = freq.get(ch, 0) + 1

max_count = max(freq.values())

for ch in freq:
    if freq[ch] == max_count:
        print(ch)
```
