```python
# Question 5: Count how many times 5 appears in (1, 5, 2, 5, 3, 5, 4, 5, 6)
num = (1, 5, 2, 5, 3, 5, 4, 5, 6)
count = 0
for i in num:
    if i == 5:
        count += 1
print(count)
```

```python

# Question 8: Check if 25 exists in tuple (10, 20, 30, 40, 50) 
num = (10, 20, 30, 40, 50)
found = False
for i in num:
    if i == 25:
        found = True
        break
if found:
    print("25 exists in tuple")
else:
    print("25 not found in tuple")
```

```python
# Question 9: Create a tuple of first 5 even numbers
num = []
for i in range(0,7,2):
    num.append(i)
num = tuple(num)
print(num)

```
