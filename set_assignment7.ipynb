```python
# Question 8: Create a set of vowels from string "hello world"
vowels = 'aeiouAEIOU'
string = "hello world"

vowel_set = set()

for i in string:
    if i in vowels:
        vowel_set.add(i)
print(vowel_set)
```

```python
# Question 9: Remove duplicates from list [1, 2, 2, 3, 4, 4, 5, 6, 6, 7] using set
nums = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
uni_set = set()
for i in nums:
    uni_set.add(i)

print(uni_set)
```
